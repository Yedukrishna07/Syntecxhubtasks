```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Fix missing values (important for plots)
df = df.dropna(subset=["Age"])

# Correct column names
column = "Age"
group = "Sex"

# Histogram
plt.figure()
df[column].hist()
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.savefig("histogram.png")
plt.show()

# KDE Plot
plt.figure()
df[column].plot(kind='kde')
plt.title("KDE Plot of Age")
plt.xlabel("Age")
plt.savefig("kde.png")
plt.show()

# Boxplot (overall)
plt.figure()
df.boxplot(column=column)
plt.title("Boxplot of Age")
plt.savefig("boxplot.png")
plt.show()

# Group KDE comparison
plt.figure()
for g in df[group].unique():
    subset = df[df[group] == g]
    subset[column].plot(kind='kde', label=f"{g}")

plt.legend()
plt.title("KDE Comparison by Sex")
plt.savefig("group_kde.png")
plt.show()

# Boxplot by group
plt.figure()
df.boxplot(column=column, by=group)
plt.title("Boxplot of Age by Sex")
plt.savefig("group_boxplot.png")
plt.show()
```
