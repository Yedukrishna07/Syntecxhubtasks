```python
import pandas as pd


# Read CSV

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

#  Basic Info

print("ORIGINAL DATA")
print(df)

print("HEAD")
print(df.head())

print("TAIL")
print(df.tail())

print("DATA TYPES ")
print(df.dtypes)


#  Statistics

print("SUMMARY STATISTICS ")

print("\nMean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nMinimum:\n", df.min(numeric_only=True))
print("\nMaximum:\n", df.max(numeric_only=True))
print("\nCount:\n", df.count())


# Filtering Rows

# Example condition: Salary > 60000
if "Salary" in df.columns:
    filtered_df = df[df["Salary"] > 60000]
else:
    filtered_df = df

print("\n===== FILTERED DATA (Salary > 60000) =====")
print(filtered_df)

# Selecting Columns

columns_to_select = df.columns[:2]
selected_df = df[columns_to_select]

print("\n===== SELECTED COLUMNS =====")
print(selected_df)


# Slicing Data

subset_df = df.iloc[:3]

print("\n===== SUBSET (FIRST 3 ROWS) =====")
print(subset_df)

# Save Results

filtered_df.to_csv("filtered_output.csv", index=False)
filtered_df.to_excel("filtered_output.xlsx", index=False)
```
